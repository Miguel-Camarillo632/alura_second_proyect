#📌 Extracción

```python
import pandas as pd
import requests

# URL de la API (ejemplo - reemplaza con la URL real proporcionada)
url = "https://raw.githubusercontent.com/ingridcristh/challenge2-data-science-LATAM/main/TelecomX_Data.json"


# 1. Extraer datos de la API
response = requests.get(url)
data = response.json()

# 2. Transformar a DataFrame
df = pd.DataFrame(data)

# Mostrar las primeras filas para verificar
print(df.head())
print("\nInformación del DataFrame:")
print(df.info())

# Opcional: Guardar en CSV para análisis posterior
df.to_csv('telecom_data.csv', index=False)
```

#🔧 Transformación

```python
import pandas as pd

# 2. Análisis inicial
print("🔍 Dimensiones del DataFrame:")
print(f"Filas: {df.shape[0]}, Columnas: {df.shape[1]}")

print("\n📌 Primeras filas del dataset:")
display(df.head(3))

print("\n🧠 Tipos de datos y valores nulos:")
print(df.info())

# 3. Estadísticas descriptivas
print("\n📊 Estadísticas descriptivas:")
display(df.describe(include='all'))

# 4. Análisis de columnas clave para evasión (churn)
print("\n🔎 Valores únicos en columna de churn (si existe):")
if 'churn' in df.columns:
    print(df['churn'].value_counts(dropna=False))
elif 'evasion' in df.columns:  # Diferentes nombres posibles
    print(df['evasion'].value_counts(dropna=False))
else:
    print("⚠️ No se encontró columna explícita de churn/evasión. Revisar diccionario.")

# 5. Identificar columnas categóricas/númericas
categorical_cols = df.select_dtypes(include=['object', 'bool']).columns
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns

print("\n📌 Columnas categóricas:", list(categorical_cols))
print("📌 Columnas numéricas:", list(numeric_cols))
```

```python
import json
import pandas as pd
from pandas import json_normalize

# 1. Desanidar las columnas que contienen diccionarios
def expand_dict_columns(row, column_name):
    try:
        if isinstance(row[column_name], str):
            return json.loads(row[column_name])
        return row[column_name]
    except:
        return {}

# Aplicar a cada columna anidada
for col in ['customer', 'phone', 'internet', 'account']:
    df[col] = df.apply(lambda x: expand_dict_columns(x, col), axis=1)

# 2. Crear DataFrames separados para cada grupo de datos
customer_df = json_normalize(df['customer'])
phone_df = json_normalize(df['phone'])
internet_df = json_normalize(df['internet'])
account_df = json_normalize(df['account'])

# 3. Combinar todo en un solo DataFrame
full_df = pd.concat([
    df[['customerID', 'Churn']],
    customer_df,
    phone_df,
    internet_df,
    account_df
], axis=1)

# 4. Limpieza post-combinación - VERSIÓN MEJORADA
# Primero reemplazar 'Yes'/'No' y luego manejar valores vacíos
full_df['Churn'] = full_df['Churn'].replace({
    'Yes': 1,
    'No': 0,
    '': pd.NA  # Convertir valores vacíos a NA
})

# Convertir a float (manejando NA)
full_df['Churn'] = pd.to_numeric(full_df['Churn'], errors='coerce')

# Eliminar registros con Churn desconocido (opcional)
full_df = full_df.dropna(subset=['Churn'])

# 5. Verificación
print("🔍 Columnas disponibles después de desanidar:")
print(full_df.columns.tolist())

print("\n📊 Distribución de Churn actualizada:")
print(full_df['Churn'].value_counts(normalize=True).map("{:.2%}".format))

print("\n💡 Muestra de datos completos:")
display(full_df.head(3))
```

```python
# 1. Calcular el valor diario (asumiendo 30 días/mes para simplificar)
full_df['Cuentas_Diarias'] = full_df['Charges.Monthly'] / 30

# 2. Redondear a 2 decimales para mejor visualización
full_df['Cuentas_Diarias'] = full_df['Cuentas_Diarias'].round(2)

# 3. Verificar resultados
print("\n🔍 Muestra de la nueva columna:")
print(full_df[['Charges.Monthly', 'Cuentas_Diarias']].head())

# 4. Análisis rápido: Comparación entre clientes que abandonan vs. permanecen
print("\n📊 Gasto diario promedio por estado de Churn:")
print(full_df.groupby('Churn')['Cuentas_Diarias'].mean().map("${:.2f}".format))
```

```python
# Seleccionar columnas numéricas
numeric_cols = full_df.select_dtypes(include=['int64', 'float64']).columns.tolist()

# Estadísticos descriptivos
desc_stats = full_df[numeric_cols].describe(percentiles=[.25, .5, .75]).transpose()
desc_stats['IQR'] = desc_stats['75%'] - desc_stats['25%']  # Rango intercuartílico

# Mostrar resultados
print("📊 Estadísticas descriptivas para variables numéricas:")
display(desc_stats[['count', 'mean', 'std', 'min', '25%', '50%', '75%', 'max', 'IQR']])

categorical_cols = ['Contract', 'PaymentMethod', 'InternetService', 'gender']

print("📊 Distribución de variables categóricas:")
for col in categorical_cols:
    if col in full_df.columns:
        print(f"\n🔹 {col}:")
        display(full_df[col].value_counts(normalize=True).map("{:.1%}".format))

if 'Churn' in full_df.columns:
    print("\n📌 Comparación por estado de Churn (0=Retenido, 1=Abandono):")

    # Para variables numéricas
    print("\n🔢 Medias numéricas por grupo:")
    display(full_df.groupby('Churn')[numeric_cols].mean().transpose())

    # Para categóricas
    print("\n🔠 Distribución categórica por grupo:")
    for col in categorical_cols:
        if col in full_df.columns:
            display(pd.crosstab(full_df['Churn'], full_df[col], normalize='index').map("{:.1%}".format))
```

```python
# Convertir Churn a categórico para mejores etiquetas
full_df['Churn_Cat'] = full_df['Churn'].map({0: 'Retenido', 1: 'Abandono'})

# Procesar columnas one-hot (si existen)
categorical_vars = {
    'gender': 'Género',
    'SeniorCitizen': 'Adulto Mayor',
    'Partner': 'Pareja',
    'Dependents': 'Dependientes',
    'Tipo_Contrato': full_df.filter(like='Tipo_Contrato_').idxmax(axis=1).str.replace('Tipo_Contrato_', ''),
    'Metodo_Pago': full_df.filter(like='Metodo_Pago_').idxmax(axis=1).str.replace('Metodo_Pago_', ''),
    'InternetService': full_df.filter(like='Servicio_Internet_').idxmax(axis=1).str.replace('Servicio_Internet_', '')
}

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from IPython.display import display

# 1. Configuración de variables categóricas
categorical_vars = {
    'Género': 'gender',
    'Adulto Mayor': 'SeniorCitizen',
    'Pareja': 'Partner',
    'Dependientes': 'Dependents',
    'Tipo Contrato': full_df.filter(like='Tipo_Contrato_').idxmax(axis=1).str.replace('Tipo_Contrato_', ''),
    'Método Pago': full_df.filter(like='Metodo_Pago_').idxmax(axis=1).str.replace('Metodo_Pago_', ''),
    'Servicio Internet': full_df.filter(like='Servicio_Internet_').idxmax(axis=1).str.replace('Servicio_Internet_', '')
}

# 2. Crear figura con tamaño dinámico
n_vars = len(categorical_vars)
n_cols = 3
n_rows = (n_vars + n_cols - 1) // n_cols  # Cálculo dinámico de filas

plt.figure(figsize=(15, 5*n_rows))

# 3. Gráficos de barras apiladas
for i, (name, var_data) in enumerate(categorical_vars.items(), 1):
    plt.subplot(n_rows, n_cols, i)

    # Calcular proporciones
    cross_tab = pd.crosstab(var_data, full_df['Churn'], normalize='index')*100

    # Ordenar por tasa de churn descendente
    cross_tab = cross_tab.sort_values(by=1, ascending=False)

    # Gráfico
    bars = cross_tab.plot(kind='bar', stacked=True,
                         color=['#3498db', '#e74c3c'],
                         ax=plt.gca())

    plt.title(f'Churn por {name}')
    plt.ylabel('Proporción (%)')
    plt.xticks(rotation=45)

    # Leyenda solo en el primer gráfico
    if i != 1:
        plt.legend().remove()
    else:
        plt.legend(['Retenido', 'Abandono'], title='Estado')

    # Etiquetas de porcentaje
    for bar in bars.containers[1]:
        height = bar.get_height()
        if height > 5:  # Solo mostrar porcentajes >5%
            plt.text(bar.get_x() + bar.get_width()/2.,
                     bar.get_y() + height/2,
                     f'{height:.1f}%',
                     ha='center', va='center',
                     color='white', fontweight='bold')

plt.tight_layout()
plt.show()

# 4. Tabla resumen de tasas de churn
print("\n📊 Tasa de Abandono por Categoría:")
churn_rates = {}
for name, var_data in categorical_vars.items():
    rate = full_df.groupby(var_data)['Churn'].mean()*100
    churn_rates[name] = rate.sort_values(ascending=False)

summary_df = pd.concat(churn_rates, axis=1)
display(summary_df.style.background_gradient(cmap='Reds', axis=None))
```
